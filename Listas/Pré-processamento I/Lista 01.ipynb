### 4 - Quais os ingredientes que aparecem nos produtos? Mostre quantos são, os que mais aparecem e os que menos aparecem.

```python
import re
```

```python
# transformo para o tipo string e copio a série
ingredientes = df['features.value'].astype('str').copy()
```

```python
ingredientes
```

```python
# quebro os elementos pela vírgula, faço virar uma lista de strings
ingredientes = ingredientes.str.split(',')
```

```python
ingredientes
```

```python
# padronizo os elementos da lista
# retiro o espaço
# coloco as iniciais em maiúsculo 
ingredientes = ingredientes.apply(lambda l : list(map(str.capitalize, map(str.strip, l))))
```

```python
ingredientes
```

```python
contagem = {}
for lista_ingredientes in ingredientes:
    for ingrediente in lista_ingredientes:
        ingrediente = re.sub('\.|\*|\)|\(|\]|\And', '', ingrediente) 
        if ingrediente in contagem.keys():
            contagem[ingrediente] += 1
        else:
            contagem[ingrediente] = 1
```

```python
contagem_de_ingredientes = pd.DataFrame({'Ingredientes' : list(contagem.keys()), 'Quantidade' : list(contagem.values())})
```

```python
contagem_de_ingredientes.shape
```

```python
contagem_de_ingredientes
```

```python
contagem_de_ingredientes = contagem_de_ingredientes[contagem_de_ingredientes.Ingredientes != '']
```

```python
print(*contagem_de_ingredientes[contagem_de_ingredientes.Quantidade == 1]['Ingredientes'].values, sep='\n')
```
