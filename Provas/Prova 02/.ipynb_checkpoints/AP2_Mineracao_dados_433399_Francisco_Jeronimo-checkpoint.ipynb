### 2) Aplique as etapas de pré-processamneto necessárias (remoção de atributos, tratamento de valores faltantes, escalonamneto/normalização de atributos, remoção de outliers). Nesta questão, indique qual a etapa está sendo aplicada e mostre, a partir dos dados, o porquê você decidiu ou não por aplicá-la.

```python
df.isnull().sum()
```

### Remoção de atributos

- **Unnamed: 32**: Coluna com todos os valores nulos
- **id**: Coluna sem ganho de informação para o agrupamento

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
```

### Encoding

É utilizada a transformação da feature categórica em numérica para que possa ser utilizada posteriormente no algoritmo DBSCAN e em redes neurais.

```python
df['diagnosis'] = df['diagnosis'].map({'B' : 0, 'M' : 1})
```

### Divisão dos dados em treino e teste

A Divisão vai ser utilizada para o treinamento das redes neurais.

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

### Escalonamento

O escalonamento é útil para evitar disparidade entre features que possuem faixas de valores diferentes, os dados são padronizados para se ter  uma escala de valores mais uniforme e sem discrepâncias que possam influenciar os algoritmos.

```python
scaler = StandardScaler()

X_train_s = scaler.fit_transform(X_train)
X_test_s =  scaler.transform(X_test)
```

### 3) Usando a biblioteca Keras, defina e treine uma rede neural de múltiplas camadas densas para resolver o problema da classifição de tumores:  

*  O modelo deve ter duas camadas ocultas
*  A camada de saída deve usar a função de ativação 'sigmoid'
*  Mostre o número de neurônios utilizado em cada camada da sua arquitetura





```python
# 20 neurônios nas camadas de entrada/ocultas
# 1 neurônio de saída

model = models.Sequential()

model.add(layers.Dense(20, input_dim=X_train_s.shape[1])) # input
model.add(layers.Dense(20)) # camadas ocultas
model.add(layers.Dense(20)) # camadas ocultas
model.add(layers.Dense(1, activation='sigmoid')) # output
```

```python
model.summary()
```

### 4) Divida o conjunto em treino, validação e teste 70/10/20. Treine o modelo usando 50 épocas batches de tamanho 20, a função de loss 'binary_crossentropy', 'accuracy' como métrica e o Gradiente Descendente Estocástico como algortimo de otimização.

```python
model.compile(optimizer='sgd',loss='binary_crossentropy', metrics=['acc'])
```

```python
m = model.fit(X_train_s, y_train, validation_split=0.1, epochs=50, batch_size=20)
```

#### a) Plot um gráfico da variação da loss para os conjuntos de validação e treino. 
Qual a melhor época para o conjunto de validação em função da loss?

**Respostas**: A partir da faixa de Epoch entre 12-20. Os valores se estabilizam e a variação é bem baixa.

```python
plt.figure(figsize=(15, 7))
plt.grid()
plt.plot(m.history['loss'], '-o',label='loss');
plt.plot(m.history['val_loss'], '-o',label='val_loss');
plt.xticks(np.arange(0, 51, 5))
plt.title('Variação loss', fontsize=15)
plt.xlabel('Epoch', fontsize=15)
plt.ylabel('loss', fontsize=15)
plt.legend();
```

#### b) Plot um gráfico da variação da acurácia para os conjuntos de validação e treino. Qual a melhor época para o conjunto de validação em função da acurácia?

**Respostas**: A Epoch 4. A faixa de epoch 4-8 possue os valores de acurácia constantes e mais altos.


Obs.: No gráfico o valor das epochs começam a partir de 0, adiciono +1 na interpretação, por isso a escolha de 16 epochs.

```python
plt.figure(figsize=(15, 7))
plt.grid()
plt.plot(m.history['acc'], '-o',label='acc');
plt.plot(m.history['val_acc'], '-o',label='val_acc');
plt.xticks(np.arange(0, 51, 5))
plt.title('Variação da Acurácia', fontsize=15)
plt.xlabel('Epoch', fontsize=15)
plt.ylabel('Acurácia', fontsize=15)
plt.legend();
```

#### c) Qual a acurácia do modelo no conjunto de teste? Como esse modelo poderia ser melhorado? Explique.

**Resposta**: Poderiamos experimentar diferentes quantidades de neurônios nas camadas e utilizar diferentes funções de ativação.

```python
loss, acc = model.evaluate(X_test_s, y_test)
```

```python
print(f'Acurácia:{acc}\nLoss {loss}')
```

### 5) Para o mesmo conjunto de dados usado nas questões anteriores, utilize apenas os atributos 'perimeter_mean' e 'area_mean'. Use o atributo de classe apenas para avaliar os resultados da clusterização.

a) Aplique o algoritmo DBSCAN, variando os valores de eps e min_samples (execute ao menos 3 combinações diferentes), mostrando um gráfico do resultado de cada clusterização. 

b) Os clusters formados foram comaptíveis com as classes? Explique.

c) Foi possível detectar outliers? Explique.

### Escalonamento

Aplico o min-max com o intuito de deixar a escala dentro do intervalo de [0, 1]. Isso vai ajudar na escolha do **eps** e **min_samples**.

```python
min_scaler = MinMaxScaler()

df.iloc[:, 1:] = min_scaler.fit_transform(df.iloc[:, 1:])
```

```python
def plot_dbscan(df, labels):
    plt.figure(figsize=(15, 7))

    sns.scatterplot(x=df.loc[:, 'perimeter_mean'], 
                    y=df.loc[:, 'area_mean'], 
                   hue=labels, 
                   palette='viridis');
```

#### Plot com as classes originais

```python
plot_dbscan(df, df['diagnosis'].values)
```

#### Execução 01

O gráfico consegue identificar alguns pontos como outliers por todo o gráfico,  os pontos que estão afastados no final da cauda no lado superior do gráfico são praticamente agrupados como um 3° cluster classificados como outliers, a região é menos densa em comparação com as demais.

```python
dbscan = DBSCAN(eps=0.5, min_samples=10)

dbscan.fit(df)

plot_dbscan(df, dbscan.labels_)
```

#### Execução 02

Os clusters se assemelham bastante com o gráfico das classes originais. Com apenas 3 pontos classificados como outliers, sendo 2 destes os pontos mais afastados do agrupamento, possuindo baixa densidade em sua volta. 

```python
dbscan = DBSCAN(eps=1, min_samples=10)

dbscan.fit(df)

plot_dbscan(df, dbscan.labels_)
```

#### Execução 03

Situação parecida com a execução 02.

```python
dbscan = DBSCAN(eps=0.950, min_samples=5)

dbscan.fit(df)

plot_dbscan(df, dbscan.labels_)
```
